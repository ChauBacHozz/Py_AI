```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.signal import savgol_filter

from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_excel("MTND3.xlsx")
df = df.drop("STT", axis = 1)
df = df.sample(frac=1)
df.head()
```

```python
df_train = df[:int(0.7 * len(df))].values
df_test = df[int(0.7 * len(df)):].values
# df_train
X_train = df_train[:,0:3]
Y_train = df_train[:,3:]
X_test = df_test[:,0:3]
Y_test = df_test[:,3:]

print(Y_test)
```

```python
n = 14
```

```python
pls = PLSRegression(n_components=n)
```

```python
pls.fit(Y_train, X_train)
X_pred = pls.predict(Y_test)
print(X_pred)
print()
print(X_test)
```

```python
ss = ((X_pred - X_test) / X_test ) * 100
ss
```
